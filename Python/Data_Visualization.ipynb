# Data Visualization — Matplotlib, Seaborn, Plotly
Includes histograms, boxplots, bar charts, pie charts, heatmaps, scatter plots, trend charts, and interactive charts with business interpretation.

```python
from pathlib import Path
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns
import plotly.express as px
sns.set_theme(style='whitegrid')
PROJECT_ROOT=Path('..') if Path('../Dataset').exists() else Path('/home/user/Data-Analytics-Project')
df=pd.read_csv(PROJECT_ROOT/'Dataset/Cleaned_Data.csv',parse_dates=['match_date'])
```

## Histogram: Runs per delivery

```python
plt.figure(figsize=(8,4)); sns.histplot(df.total_runs,bins=8,color='#2563eb'); plt.title('Distribution of Runs per Delivery'); plt.show()
```

Most deliveries create low runs; boundaries and extras drive the high-impact tail.

## Boxplot: Phase performance

```python
plt.figure(figsize=(10,5)); sns.boxplot(data=df,x='phase',y='total_runs'); plt.title('Runs by Match Phase'); plt.show()
```

Phase-level dispersion helps identify acceleration and risk windows.

## Bar chart: Top teams

```python
top=df.groupby('batting_team').total_runs.sum().sort_values(ascending=False).head(10); plt.figure(figsize=(12,5)); sns.barplot(x=top.values,y=top.index); plt.title('Top Batting Teams by Runs'); plt.show()
```

## Pie chart: Run share by phase

```python
phase=df.groupby('phase').total_runs.sum().sort_values(ascending=False); plt.figure(figsize=(6,6)); plt.pie(phase.values,labels=phase.index,autopct='%1.1f%%'); plt.title('Run Share by Phase'); plt.show()
```

## Heatmap: Numeric relationships

```python
cols=['over_number','batsman_runs','extra_runs','total_runs','is_wicket','legal_ball','is_boundary','is_dot_ball','is_extra']; plt.figure(figsize=(9,6)); sns.heatmap(df[cols].corr(),annot=True,cmap='coolwarm',fmt='.2f'); plt.title('Correlation Heatmap'); plt.show()
```

## Scatter: Player volume vs strike rate

```python
bat=df.groupby('batter').agg(runs=('batsman_runs','sum'),balls=('legal_ball','sum'),sixes=('is_six','sum')).reset_index(); bat=bat[bat.balls>=200]; bat['strike_rate']=bat.runs/bat.balls*100; plt.figure(figsize=(10,6)); sns.scatterplot(data=bat,x='balls',y='strike_rate',size='runs',hue='sixes'); plt.title('Batter Volume vs Strike Rate'); plt.show()
```

## Interactive trend chart

```python
season=df.groupby('match_year').agg(runs=('total_runs','sum'),balls=('legal_ball','sum')).reset_index(); season['run_rate']=season.runs/(season.balls/6); px.line(season,x='match_year',y='run_rate',markers=True,title='Interactive Season Run Rate Trend').show()
```

## Interactive regional chart

```python
city=df.groupby('venue_city').agg(matches=('match_id','nunique'),runs=('total_runs','sum'),wickets=('is_wicket','sum')).reset_index().sort_values('matches',ascending=False).head(20); px.bar(city,x='venue_city',y='runs',color='matches',hover_data=['wickets'],title='Venue-City Performance').show()
```
